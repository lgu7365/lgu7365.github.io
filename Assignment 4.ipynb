2) 파일 s.txt을 읽어서 각 라인에 있는 첫 번째 단어(문자열) 자체들을 기준으로 라인별 정렬후 파일 s1.txt에 그 결과를 기록하는 코드를 작성하시오. 즉, 프로그램 수행 후 s1.txt에 있는 파일 내용은 다음과 같다.

3) 파일 s.txt을 읽어서 각 라인에 있는 두 번째 단어(문자열) 자체들을 기준으로 라인별 정렬후 파일 s2.txt에 그 결과를 기록하는 코드를 작성하시오. 즉, 프로그램 수행 후 s2.txt에 있는 파일 내용은 다음과 같다.
> ham bird  
cat dog  
pig ham  
dog pig

```python
f = open("s.txt", 'r')
list = f.readlines()
list1 = sorted(list, key=lambda x : x[1])
f.close()
f1 = open("s2.txt", 'w')
f1.writelines(list1)
f1.close()

f2 = open("s2.txt", 'r')
print(f2.read())
f1.close()
```

4) 파일 s.txt을 읽어서 각 라인들에 있는 모든 단어들을 순차적으로 다시 나열하되 각 라인에 세 개의 단어들이 오도록 하여 s3.txt에 기록하는 코드를 작성하시오. 즉, 프로그램 수행 후 s3.txt에 있는 파일 내용은 다음과 같다.
> pig ham cat  
dog ham bird  
dog pig

```python
f = open("s.txt", 'r')
list = f.readlines()
a = []
for i in range(len(list)):
    list1 = list[i].split()
    for j in range(len(list1)):
        a.append(list1[j])
f.close()
f1 = open("s3.txt", 'w')
for i in range(len(a)):
    f1.write(a[i]+' ')
    if(i%3 == 2):
        f1.write('\n')
f1.close()

f2 = open("s3.txt", 'r')
print(f2.read())
f1.close()
```

해결방법
1) open()함수를 사용해 'w'모드로 연 뒤 다음 내용을 write()함수를 통해 써줍니다.  
2) s.txt를 'r'모드로 연 뒤 s의 모든 내용을 readlines()함수를 사용해 리스트로 받아준 뒤 sort()로 정렬하고 정렬된 리스트를 s1.txt에 넣어줍니다.  
3) s.txt를 'r'모드로 연 뒤 s의 모든 내용을 readlines()함수를 사용해 리스트로 받아준 뒤 sorted()함수의 key를 이용해 두번째 단어를 기준으로 정렬해줍니다. 그리고 정렬된 리스트를 s2.txt에 넣어줍니다.  
4) s.txt를 'r'모드로 연 뒤 s의 모든 내용을 readlines()함수를 사용해 리스트로 받아준 뒤 빈 리스트 a를 생성하고 list를 ' '을 기준으로 짤라준 뒤 a에 넣어줍니다. 그리고 for문으로 s3.txt에 a의 값들을 차례대로 ' '와 함께 써주고 3번째 단어마다 '\n'을 넣어줘 각 라인에 세 개의 단어가 오도록 합니다.

## 3. range() 함수와 유사한 frange() 함수를 다음 조건을 참고하여 만드시오.

+ 1) frange()의 인자 구성은 다음과 같이 range와 동일하지만 각 인수들은 음수를 받지 않는다고 가정한다.
> range(stop)  
range(start, stop[, step])

+ 2) frange() 함수의 인자에 대한 기본 시작(start) 값은 0.0이고, 기본 단계(step) 값은 0.1이다.
+ 3) frange 사용 예
    * 3-1) frange(0.5)
        - [0.0, 0.1, 0.2, 0.3, 0.4]
    * 3-2) frange(1.0, 2.0)
        - [1.0, 1.1, 1.2, 1.3, 1.4, 1.5, 1.6, 1.7, 1.8, 1.9]
    * 3-3) frange(2.2, 4.0, 0.5)
        - [2.2, 2.7, 3.2, 3.7]

```python
def frange(start = None, stop = None, step=0.1):
    a = []
    if(start != None and stop == None):
        stop = start
        start = 0.0
    while(start < stop):
        a.append(round(start,10))
        start += step
    print(a)
    
frange(0.5)
frange(1.0, 2.0)
frange(2.2, 4.0, 0.5)
```

1. 먼저 start,stop,step에 기본값을 설정해 줍니다. 
2. 값들을 저장할 빈 리스트 a를 생성해줍니다.
3. 인자의 값을 하나만 받을 때를 구분하여 start만 값을 받고 stop은 값을 받지 않았을 경우 stop에 받은 값을 넣어주고 start는 0.0으로 하여 기본값을 설정해 줍니다.
4. while문을 통해 start의 값이 stop을 넘지 않을 때까지 반복해 a에 start의 값을 넣어주고 start를 step만큼 증가시켜줍니다.
5. 그냥 float값을 넣는 경우 근사값을 갖기 때문에 이 문제를 해결하기 위해 round함수를 사용해 반올림을 해주었습니다.

## 5. 여러 단어로 이루어진 문자열을 입력받아 각 단어의 첫글자로 이루어진 단어를 대문자로 출력하는 myinitial() 함수를 다음 조건을 참고하여 작성하시오.

+ 1) 다음에 제시되는 함수들을 모두 이용해야 한다.

    - split
    - map
    - join
+ 2) myinitial() 함수 사용 예

    - myinitial("as soon as possible")
        * ASAP

```python
def myinitial(s):
    a = s.split()
    b = list(map(lambda x: x[0], a))
    result = "".join(b)
    result = result.upper()
    print(result)
myinitial("as soon as possible")
```

1. 먼저 문자열 s를 입력받아 각 단어를 나누어 a에 저장해줍니다.
2. map함수와 lambda함수를 이용하여 a의 값들을 맨 앞 글자 하나 씩만 b에 저장하여 줍니다.
3. join함수를 이용하여 리스트를 문자열로 변환해 result에 저장해줍니다.
4. upper함수를 이용해 result를 대문자로 바꾸어 출력해줍니다.

## 6. 음이 아닌 정수 n를 입력받으면 n! (factorial)을 계산하는 myfact() 함수를 재귀적 함수로 구현하시오.
+ [옵션] 참을 수 있는 정도 만큼의 수행시간을 직접 기다려보면서 n을 늘려보도록 합시다.

```python
def myfact(n):
    if n < 0:
        print("음이 아닌 정수를 입력하세요")
    elif n == 0:
        return 1
    else:
        return n * myfac(n-1)
myfact(2700)
```

n! = n*(n-1)!이므로 myfact(n) = n * myfact(n-1) 로 재귀 함수를 구현합니다. 0! = 1 이므로 n이 0이 될 때까지 반복하여 주고 0인 경우 1을 반환해 줍니다.
